# Biofeedback
## Get to Know Biofeedback
이완훈련(안정화, relaxation) + 신체반응 측정 --> 치료 & **자가 훈련**


왜 신체반응을 측정하는가?
정신활동(생각, 감정 등) -- 신체활동 (호흡, 땀 등)
직접, 즉시 측정 불가      대신 측정, 관찰


어떤 것을 측정하는가?
자율신경계 (호흡, **심장박동**, 땀, 체온, **신경근육** 등)


자율신경계는 신체활동 & 정신활동에 영향을 받는다.
ex) 자기 전 안좋은 생각 - 두근두근


어떤 훈련인가?
불수의적인 자율신경계의 반응을 훈련을 통해 수의적/의식적으로 조절하여 안정화하는 것
